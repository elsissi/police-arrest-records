<h2>Data Cleaning</h2>

```python
data.drop(["ID Reference Number "], axis=1, inplace=True)
```

so handling missing values

```python
def uniquevalues(data) :
    for i in data.columns:
        print(i, ":", len(data[i].unique()))
print("unique values :-")
uniquevalues(data)
```

```python
def nullvalues(data):
    return DataFrame({col: data[col].isnull().sum() for col in data.columns}, index=["null values"]).T
print(data.shape)
nullvalues(data)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

<h2>Handling Missing Values</h2>

```python
# for missing values in district of occurrence
data["District of occurrence"].value_counts().head(10)
```

```python
def fill_missing_random_with_prob(data, value, prob,column):
    missing_indices = data[column].isna()
    random_values = np.random.choice(value, size=missing_indices.sum(), p= prob)
    data.loc[missing_indices, column] = random_values
    return data

```

```python
values1 = [6.0,3.0,4.0,5.0,1.0,2.0,8.0]
prop1 = list(data["District of occurrence"].value_counts().head(10)/(data.shape[0]-data["District of occurrence"].isnull().sum()))
data = fill_missing_random_with_prob(data, values1, prop1,"District of occurrence")
```

```python
# for missing values in assigned division
data["Assigned Division"].value_counts().head(30)
```

```python
values2 = ['3D','4D','5D','GPD','2D','6D','RCPD','TOD','1D','TPPD']
prop2 = list((data["Assigned Division"].value_counts().head(10)/(data.shape[0]-data["Assigned Division"].isnull().sum()))*(data["Assigned Division"].value_counts().sum()/data["Assigned Division"].value_counts().head(10).sum()))
data = fill_missing_random_with_prob(data, values2, prop2,"Assigned Division")
```

```python
# for missing values in assigned bureau
data["Assigned Bureau"].value_counts().head()
```

```python
values3 = ['PSB','FSB','ISB','MSB']
prop3 = list(data["Assigned Bureau"].value_counts().head(4)/(data.shape[0]-data["Assigned Bureau"].isnull().sum()))

data = fill_missing_random_with_prob(data, values3, prop3,"Assigned Bureau")
```
